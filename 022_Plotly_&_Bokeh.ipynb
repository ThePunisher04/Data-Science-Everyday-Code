```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[3,4,5,6,7], mode="markers"))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[3,14,55,60,7], mode="lines"))
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip, mode="markers"))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5], y=[3,14,55,60,7]))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=tips.total_bill))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip, mode="markers", marker_size = tips['size']*5))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill, y=tips.tip, mode="markers", marker_size = tips['size']*5, z=tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=[3,5,7,8,1], y=[40,32,66,77,11], mode="lines", marker_size = tips['size']*5, z=[9,6,3,7,5]))
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook #print the graph in notebook
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')
p = figure(title = 'flower test')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

image.png

```python
output_file('test_line.html')
p = figure(title = 'flower test LINE')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.line(flowers['petal_length'], flowers['petal_width'])
show(p)
```
