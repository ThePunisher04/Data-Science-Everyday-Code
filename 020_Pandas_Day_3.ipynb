#Merge and Join

```python
df1 = pd.DataFrame(
    {
        'key1' : [1,2,4,5,6],
        'key2' : [4,5,6,7,8],
        'key3' : [3,4,5,6,6]
    }
)
```

```python
df1
```

```python
df2 = pd.DataFrame(
    {
        'key1' : [1,2,45,6,67],
        'key4' : [56,5,6,7,8],
        'key5' : [3,56,5,6,6]
    }
)

df2
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='cross')
```

```python
pd.merge(df1, df2, how='left', left_on = 'key2', right_on = 'key4')
```

```python
df1 = pd.DataFrame(
    {
        'key1' : [1,2,4,5,6],
        'key2' : [4,5,6,7,8],
        'key3' : [3,4,5,6,6]
    },
    index = ['a','b','c','d','e']
)
```

```python
df1
```

```python
df2 = pd.DataFrame(
    {
        'key6' : [1,2,45,6,67],
        'key4' : [56,5,6,7,8],
        'key5' : [3,56,5,6,6]
    },
    index = ['a','b','h','i','j']
)

df2
```

```python
df1.join(df2, how = 'left')
#join happens on the common indexes
#column name should not be same
```

```python
df1.join(df2, how = 'right')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1.join(df2, how = 'cross')
```

```python
df1.join(df2, how = 'inner')
```
