```python
plt.figure()

plt.subplot(2, 3, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 3, 2)
plt.plot(x, np.cos(x))

plt.subplot(2, 3, 3)
plt.plot(x, np.tan(x))

plt.subplot(2, 3, 4)
plt.plot(x, 1/np.sin(x))

plt.subplot(2, 3, 5)
plt.plot(x, 1/np.cos(x))

plt.subplot(2, 3, 6)
plt.plot(x, 1/np.tan(x))

plt.show()
```

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'green')
plt.plot(x, np.sin(x-1), color = 'blue')
plt.plot(x, np.sin(x-2), color = 'red')
plt.plot(x, np.sin(x-3), color = (1.0, 0.5, 0))
```

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')
plt.title("This is sin(x) plot")
plt.grid()
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'blue', label = 'Sinx')
plt.plot(x, np.cos(x), color = 'red', label = 'Cosx')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')
plt.title("This is sin(x) vs cos(x) plot")
plt.grid()
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.legend()
plt.show()
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

plt.figure(figsize = (6, 2))
plt.scatter(x, y, c = 'r', alpha = .9, marker = 'X')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Scatter Plot')
plt.show()
```
